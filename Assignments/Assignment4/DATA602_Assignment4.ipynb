## Question 1

Create a class called BankAccount that has four attributes: bankname, firstname,
lastname, and balance.

The default balance should be set to 0.

In addition create:
* A method called deposit() that allows the user to make deposits into their balance.
* A method called withdrawal() that allows the user to withdraw from their balance.
* Withdrawal may not exceed the available balance. Hint: consider a conditional argument in your withdrawal() method.
* Use the \_\_str\_\_() method in order to display the bank name, owner name, and current balance.
* Make a series of deposits and withdrawals to test your class.

```python
class BankAccount:
    def __init__(self, bankname, firstname, lastname, balance = 0):
        self.bankname = bankname
        self.firstname = firstname
        self.lastname = lastname
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        
    def withdrawal(self, amount):
        if(amount <= self.balance):
            self.balance -= amount
        else:
            print("Unsuccessful withdrawl: Withdrawl exceeds account balance. No money will be taken out of account. Please enter in a withdrawal amount that is equal to or less than account balance. \n")
            
    def __str__(self):
        return 'Bank Name: {self.bankname}\nOwner Name: {self.firstname} {self.lastname}\nCurrent Balance: {self.balance}'.format(self=self)
    
# Testing

print('Test 1 \n')
account1 = BankAccount('Chase', 'Charles', 'Longbottom')
account1.deposit(100)
account1.withdrawal(25)
account1.deposit(1050)
account1.deposit(466)
account1.withdrawal(900)
account1.withdrawal(42)
print(account1)


print('\nTest 2 \n')
account1 = BankAccount('Bank of America', 'Lauren', 'Cliffton', 1000)
account1.deposit(500)
account1.withdrawal(2000)
print(account1)
```

# Question 2

Create a class Box that has attributes length and width that takes values for length
and width upon construction (instantiation via the constructor).

In addition, create:
* A method called render() that prints out to the screen a box made with asterisks of length and width dimensions
* A method called invert() that switches length and width with each other
* Methods get_area() and get_perimeter() that return appropriate geometric calculations
* A method called double() that doubles the size of the box. Hint: Pay attention to return value here.
* Implement \_\_eq\_\_ so that two boxes can be compared using ==. Two boxes are equal if their respective lengths and widths are identical.
* A method print_dim() that prints to screen the length and width details of the box
* A method get_dim() that returns a tuple containing the length and width of the box
* A method combine() that takes another box as an argument and increases the length and width by the dimensions of the box passed in
* A method get_hypot() that finds the length of the diagonal that cuts through the middle
* Instantiate 3 boxes of dimensions 5,10 , 3,4 and 5,10 and assign to variables box1, box2 and box3 respectively
* Print dimension info for each using print_dim()
* Evaluate if box1 == box2, and also evaluate if box1 == box3, print True or False to the screen accordingly
* Combine box3 into box1 (i.e. box1.combine())
* Double the size of box2
* Combine box2 into box1

```python
from math import sqrt

class Box:
    def __init__(self, length, width):
        self.length = length
        self.width  = width
    
    def render(self):
        for i in range(0,self.width):
            print('* ' * self.length)
    
    def invert(self):
        print("Original Length: ", self.length, "\nOriginal Width: ", self.width)
        length = self.width
        width = self.length
        self.width = width
        self.length = length
        print("Inverted Length: ", self.length, "\nInverted Width: ", self.width)
        
    def get_area(self):
        return self.length * self.width
    
    def get_perimeter(self):
        return self.length * 2 + self.width * 2
    
    def double(self):
        print("Original Size: ", self.length, " by ", self.width)
        length = self.length * 2
        width = self.width * 2
        self.length = length
        self.width = width
        print("Doubled Size: ", self.length, " by ", self.width)
        
    def __eq__(self, other):
        return self.length == other.length, self.width == other.width  
    
    def print_dim(self):
        print("Length: ", self.length, "  Width: ", self.width)
        
    def get_dim(self):
        return self.length, self.width
    
    def combine(self, other):
        self.length += other.length
        self.width += other.width
        
    def get_hypot(self):
        # a^2 + b^2 = c^2
        hypot = sqrt(self.length**2 + self.width**2)
        return hypot
        
# Testing

# Initialize box 1,2,3
box1 = Box(5,10)
box2 = Box(3,4)
box3 = Box(5,10)

# Print box dimensions
print('Box 1 Dimensions:')
box1.print_dim()
box1.render()

print('Box 2 Dimensions:')
box2.print_dim()
box2.render()

print('Box 3 Dimensions:')
box3.print_dim()
box3.render()


# Evaluate if box1 == box2
print('\n')
print('Evaluate if box1 == box2')
print(box1 == box2)

# Evaluate if box1 == box3
print('Evaluate if box1 == box3')
print(box1 == box3)

# Combine box3 into box1 (i.e. box1.combine())
print('\n')
print('Combine box3 into box1')
box1.combine(box3)
print('Combined Length: ', box1.length, '  Combined Width: ', box1.width)


# Double size of box2
print('\n')
print('Double the size of Box 2:')
box2.double()

# Combine box2 into box1
print('\n')
print('Combine box2 into box1')
box1.combine(box2)
print('Combined Length: ', box1.length, '  Combined Width: ', box1.width)
```
